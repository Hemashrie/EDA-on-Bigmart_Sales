```python
categorical_column = df.select_dtypes(include = "object").columns  ### to create a subset of categorical features
numerical_columns = df.select_dtypes(include = ["int64","float64"]) ### to create a subset of numerical features
```

```python
for cat in categorical_column:
    print(f"Unique values and their counts in {cat}")
    print(df[cat].value_counts())
    print("**************************")
```

```python
### Data in column fat content has to be fixed: Covert LF and Low fat to Low fat, and reg to Regular.
df['Item_Fat_Content']= df['Item_Fat_Content'].replace({"reg" : "Regular", "LF" : "Low Fat", "low fat" : "Low Fat"})
```

```python
df['Item_Fat_Content'].value_counts()
```

```python
### Combine all supermarket types under one supermarket Label
df['Outlet_Type'] = df['Outlet_Type'].replace({"Supermarket Type1" : "Supermarket",
                                               "Supermarket Type2" : "Supermarket",
                                               "Supermarket Type3" : "Supermarket"})
```

```python
df['Outlet_Type'].value_counts()
```

```python
for num in numerical_columns:
    print(num)
    print(df[num].unique())
```

### Step 3 : Handle Missing Data

```python
### Let us use median imputation for item_weight and imputation for outer_size.
df.fillna({'Item_Weight' : df['Item_Weight'].median(),"Outlet_Size" : df["Outlet_Size"].mode()[0]}, inplace = True)
```

```python
df.isnull().sum()
```

### step 4 : Perform Univariate Analysis with distribution plots/countplots
- Univariate analysis is the simplest form of data analysis. It involves analysing data with only one variable to understand its distribution, central tendency, and dispersion

```python
### Distribution plots and numerical columns
for num in numerical_columns:
    sns.displot(data = df, x = num, bins = 15, height = 3, aspect = 2)
    plt.title(f'Distribution plot of {num}')
    plt.show()
```

```python
for num in numerical_columns:
    if num != "Outlet_Establishment_Year":
       sns.displot(data = df, x = num, bins = 15, height = 3, aspect = 2, kde = True)
       plt.title(f'Distribution plot of {num}')
       plt.show()
```

```python
for num in numerical_columns:
    if num != "Outlet_Establishment_Year":
       sns.displot(data = df, x = num, bins = 15, height = 3, aspect = 2, kde = True)
       plt.axvline (df[num].mean(),label = "mean", c = "red")
       plt.title(f'Distribution plot of {num}')
       plt.show()
```

```python
for num in numerical_columns:
    if num != "Outlet_Establishment_Year":
       sns.displot(data = df, x = num, bins = 15, height = 3, aspect = 2, kde = True)
       plt.axvline (df[num].mean(),label = "mean", c = "red")
       plt.axvline (df[num].median(),label = "median", c = "green")
       plt.title(f'Distribution plot of {num}')
       plt.show()
```

- Item outlet sales and Item visibility is positively skewed.
- Not a clear pattern is visible in item_weight and item_mrp, the data seems symmetrical in item_weight.
- Majority of the item_weight are around 12 units.
- MRP values seems to follow approx.uniform distribution.

```python
### Create countplot for all categorical columns including establishment year.
for cat in categorical_column:
    sns.countplot(data = df, x = cat, hue = cat)
    plt.title(f"count plots of {cat}")
    plt.show()
```
